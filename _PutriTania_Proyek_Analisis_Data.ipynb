```python
# Membaca dataset
day_df = pd.read_csv('day.csv')

# Mengelompokkan data berdasarkan kondisi cuaca
grouped_data = day_df.groupby('weathersit')

# Menghitung jumlah total peminjaman sepeda untuk setiap kondisi cuaca
weather_rentals = grouped_data['cnt'].sum()

# Menyimpan hasil perhitungan dalam struktur data baru atau menampilkan hasil
print(weather_rentals)
```

```python
# Membaca dataset
day_df = pd.read_csv('day.csv')

# Mengelompokkan data berdasarkan suhu ke dalam kategori tertentu
# Misalnya, kategori bisa berupa rentang suhu tertentu (dingin, sejuk, hangat, panas)
def categorize_temperature(temp):
    if temp < 10:
        return 'Dingin'
    elif 10 <= temp < 20:
        return 'Sejuk'
    elif 20 <= temp < 30:
        return 'Hangat'
    else:
        return 'Panas'

day_df['temperature_category'] = day_df['temp'].apply(categorize_temperature)

# Menghitung jumlah total peminjaman sepeda untuk setiap kategori suhu
temperature_rentals = day_df.groupby('temperature_category')['cnt'].sum()

# Menyimpan hasil perhitungan dalam struktur data baru atau menampilkan hasil
print(temperature_rentals)
```

## Visualization & Explanatory Analysis

# Pada bulan apa paling banyak yang merental sepeda?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setel gaya plot seaborn
sns.set(style="whitegrid")

# Visualisasi tren jumlah peminjaman per bulan
plt.figure(figsize=(12, 6))
sns.lineplot(x='mnth', y='cnt', data=day_df)
plt.title('Tren Jumlah Peminjaman Sepeda per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Visualisasi data jumlah peminjaman sepeda per bulan dalam bentuk diagram batang
plt.figure(figsize=(12, 6))
plt.bar(monthly_rentals.index, monthly_rentals.values, color='blue')
plt.title('Jumlah Peminjaman Sepeda per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peminjaman')
plt.xticks(monthly_rentals.index)  # Menampilkan label bulan
plt.grid(axis='y')  # Menampilkan garis grid pada sumbu y
plt.show()
```

### Pertanyaan 2: Apakah terdapat hubungan antara suhu/kelembapan dengan jumlah peminjaman?

```python
# Visualisasi hubungan antara kondisi cuaca dan jumlah peminjaman sepeda
plt.figure(figsize=(12, 6))
sns.boxplot(x='weathersit', y='cnt', data=day_df)
plt.title('Hubungan antara Kondisi Cuaca dan Jumlah Peminjaman Sepeda')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.show()
```

```python
# Analisis korelasi dengan variabel lain jika diperlukan
correlation_matrix = day_df[['weathersit', 'temp', 'hum', 'windspeed', 'cnt']].corr()
```

```python
# Visualisasi matriks korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi antara Suhu, Kelembapan, dan Jumlah Peminjaman Sepeda')
plt.show()
```
